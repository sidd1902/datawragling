```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #iporting this for some visul representaion
```

```python
df = pd.read_csv('d1.csv')   # i have named dataset1 d1 and d2&d3 for dataset 2&3 downloded from project docs
```

```python
df
```

```python
unique_vals = df["dteday"].unique()
print(f"Unique values in 'column_name': {unique_vals}")
```

```python
unique_vals = df["season"].unique()
print(f"Unique values in 'column_name': {unique_vals}")
```

```python
unique_vals = df["hr"].unique()
print(f"Unique values in 'column_name': {unique_vals}")
```

```python
for i in df.columns:  # cheacking unique values in each cols is time consuming and makes code lenthy
    unique_vals = df[i].unique()  # so i m adding loop for this making it faster i will also do same for other opretions from now on
    print(f"Unique values in '{i}': {unique_vals}")
    print ('done') # this done is just for some gap making it more readable
```

```python
col_drop = ["yr", "mnth"] # yr and mnth is not that IMP as ther is only 0 and 1 and also this cols anr not mentioned in project docs
df.drop(columns=col_drop, inplace=True)
print(df.head())
```

```python
for i in df.columns: #if ther is any null/NaN values it will put 0 in their 
    df.fillna(value={"i": 0}, inplace=True) 
    print(f"Missing values treated in",i)
```

```python
print(df.dtypes)   #identifing datatypes
```

```python
df.hist(figsize=(20, 10))#some visule representaion
plt.suptitle("Histograms of Numeric Columns")
plt.tight_layout()
plt.show()
```

```python
df.plot(kind='box', subplots=True, layout=(1, len(df.select_dtypes(include=np.number).columns)), figsize=(10, 5), sharey=False)
plt.suptitle("Boxplots of Numeric Columns")
plt.tight_layout()
plt.show() #boxplots helps to identifi outlyers visualy
```

```python
import seaborn as sns
plt.figure(figsize=(6, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```
