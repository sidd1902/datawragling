```python
#as youn can see their are some outliers in some colunms 
# Handle outliers using the IQR method for all numeric columns
numeric_cols = df2.select_dtypes(include=np.number).columns
for col in numeric_cols: #instid of handeling one by one i m adding loop here
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Cap outliers
    df2[col] = np.where(df2[col] < lower_bound, lower_bound, np.where(df2[col] > upper_bound, upper_bound, df2[col]))
```

```python
df2.plot(kind='box', subplots=True, layout=(1, len(df.select_dtypes(include=np.number).columns)), figsize=(20, 5), sharey=False)
plt.suptitle("Boxplots of Numeric Columns")
plt.tight_layout()
plt.show()             #lets see the dff
```
